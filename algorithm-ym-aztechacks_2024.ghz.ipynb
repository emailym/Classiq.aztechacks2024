### `GHZ`
**What it computes:**
- Computes GHZ state for 3 qubits.

**Mathematical description:**
- for 3 qubits generates state 111 (2^3 -1 =7) or state 000

**Parameters:**
- none

**Result:**
- should be 7 (for 000) or 0 (for 000).

---

```python
from classiq import X


@qfunc
def main():
    q1 = QBit("q1")
    q2 = QBit("q2")
    q3 = QBit("q3")
    allocate(1, q1)
    allocate(1, q2)
    allocate(1, q3)

    H(q1)
    CX(q1,q2)
    CX(q2,q3)

# Create the quantum model
qmod = create_model(main)

from classiq import execute, synthesize, write_qmod

# Write the quantum model to a file
write_qmod(qmod, "GHZ_example")

# Synthesize the quantum program
qprog = synthesize(qmod)

# Execute the quantum program and get the result
result = execute(qprog).result()[0].value
print(result.parsed_counts)
```
